```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('matches.csv')
df.info()
```

```python
df.drop(['best_of', 'draw_size','loser_entry','loser_ht','loser_id','loser_rank_points',
         'loser_seed','match_num','minutes','score','tourney_date','tourney_id','winner_id',
         'winner_rank_points','Unnamed: 32'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

##### omplim les edats que falten amb la mitja

```python
df['winner_age'].fillna(df['winner_age'].mean(), inplace=True)
df['loser_age'].fillna(df['loser_age'].mean(), inplace=True)
```

```python
df = df.astype({'loser_age': int,'winner_age': int})
```

```python
df
```

```python
for num in df['year'].index[df['year'].apply(np.isnan)]:
    print(num)
df[np.isnan(df.year)]
```

```python
df.dropna(subset=['year'],inplace=True)
```

```python
df.to_csv('tennis.csv')
```
